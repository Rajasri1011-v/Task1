## Data cleaning

### Subtask:
Remove outliers from the numerical features of the Titanic dataset based on the boxplot visualizations.


**Reasoning**:
Remove outliers from the numerical features 'Age' and 'Fare' using the IQR method, based on the boxplots generated in the previous step.  These features exhibited clear outliers.  Create a new DataFrame 'titanic_df_no_outliers' containing the cleaned data.



```python
import pandas as pd

# Identify features with outliers (based on visual inspection of boxplots)
features_with_outliers = ['Age', 'Fare']

# Create a copy to avoid modifying the original DataFrame
titanic_df_no_outliers = titanic_df.copy()

outlier_removal_summary = []

for feature in features_with_outliers:
    Q1 = titanic_df[feature].quantile(0.25)
    Q3 = titanic_df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers before removal
    outliers_count = titanic_df[(titanic_df[feature] < lower_bound) | (titanic_df[feature] > upper_bound)].shape[0]

    # Remove outliers
    titanic_df_no_outliers = titanic_df_no_outliers[
        (titanic_df_no_outliers[feature] >= lower_bound) & (titanic_df_no_outliers[feature] <= upper_bound)
    ]

    outlier_removal_summary.append({
        "feature": feature,
        "outliers_removed": outliers_count,
        "lower_bound": lower_bound,
        "upper_bound": upper_bound
    })

# Print summary of outlier removal
for summary in outlier_removal_summary:
    print(f"Feature: {summary['feature']}")
    print(f"  Outliers removed: {summary['outliers_removed']}")
    print(f"  Lower bound: {summary['lower_bound']}")
    print(f"  Upper bound: {summary['upper_bound']}")
    print("-" * 20)

print("\nShape of the DataFrame after outlier removal:", titanic_df_no_outliers.shape)
```
